# Exploratory Data Analysis on Student Dataset

This analysis explores factors affecting student academic performance. The goal is to understand patterns, distributions, and relationships between variables.







```python
import pandas as pd

df = pd.read_csv("StudentsPerformance_with_headers.csv")
df.head()
```

# Dataset Overview
 The dataset contains demographic, academic, and behavioral information about students. It includes variables such as gender, study habits, attendance, and final grades.


```python
df.info()
```

```python
df.describe()
```

# Data Structure

The dataset contains both categorical and numerical variables. The numerical variables provide measurable insights such as grades, while categorical variables describe student characteristics.

```python
df.isnull().sum()
```

# Distribution of Grades

The histogram shows that most students are concentrated in the lower to mid grade levels, particularly between grades 1 and 3. Higher grades appear less frequently, indicating that fewer students achieve top performance. This suggests that the overall academic performance leans toward moderate levels.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['GRADE'], bins=10)
plt.title("Distribution of Student Grades")
plt.show()
```

# Grades by Gender

The boxplot shows a noticeable difference in grade distribution between the two gender groups. Students in group 2 tend to have higher median grades and a wider range of higher values compared to group 1. This suggests that group 2 generally performs slightly better academically.

```python
sns.boxplot(x='Sex', y='GRADE', data=df)
plt.show()
```

# Correlation Analysis

The correlation matrix highlights relationships between numerical variables. Most correlations are weak, suggesting that student performance is influenced by multiple factors.

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

# Conclusion

In this analysis, I explored the student dataset to understand how grades are distributed and how different factors relate to academic performance. I found that most students fall within the lower to middle grade ranges, which shows that overall performance is moderate.

When comparing grades by gender, I noticed that one group tends to have slightly higher grades, but the difference is not very large. The correlation analysis also showed that there is no single factor that strongly affects grades, which means student performance depends on multiple aspects.

Overall, this analysis helped me better understand the dataset and how exploratory data analysis can be used to identify patterns and insights.
